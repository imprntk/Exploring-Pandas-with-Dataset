```python
#we can give conditions by using lambda or by creating function 

def categorize_athelet(row):
    if row['height_cm'] < 175 and row['weight_kg'] < 70:
        return "Lightweight"
    elif row['height_cm'] < 185 or row['weight_kg'] <= 80:
        return "Middlwweight"
    else:
        return "Heavyweight"
    
bios['category'] = bios.apply(categorize_athelet, axis=1)  # row=1,column=0
```

Advanced Functionality (.shift() .rank() .rolling() .cumsum())

```python
#shift function
#means like how ur revenue doing compaired to day before soemthing like that

coffee_new['yesterday_revenu'] = coffee['revenu'].shift(2)  # 2 is gap like moving average
```

```python
coffee_new.head()
```

```python
coffee_new['pct_change'] = coffee_new['revenu'] / coffee_new['yesterday_revenu'] * 100
```

```python
coffee_new.head()
```

```python
bios['height_rank'] = bios['height_cm'].rank(ascending=False)
```

```python
bios.sort_values(['height_rank']).sample(10)[['name','height_cm']]
```

```python
#cumsum is cummulative sum
coffee_new['cumulative_revenu'] = coffee_new['revenu'].cumsum()
```

```python
coffee_new.head()
```

```python
#rolling is moving average
#rolling(3) means 3 days moving average
latte = coffee_new[coffee['Coffee Type']=='Latte'].copy()

latte['3day'] = latte['Units Sold'].rolling(3).sum()
```
